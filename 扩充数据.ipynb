```python
'''
Author: xv_rong
Date: 2022-04-07 11:20:48
LastEditors: xv_rong
'''
import numpy as np
import copy
import pandas as pd

# 读取../data下的demand.csv文件, 第一列为index, 第一行作为columns
df = pd.read_csv('./data/demand.csv', index_col=0, header=0)

# 将index转换为日期格式
df.index = pd.to_datetime(df.index)

i = len(df.groupby(df.index))
df_group = df.groupby(df.index)
while i < 8920:
    m_time_num = i
    for key, value in df_group:
        value.index = value.index + pd.Timedelta(minutes=5 * m_time_num)
        df = df.append(value)
        i += 1
        if i == 8928:
            break
```

```python
# 将df 的index 转换为 yyyy-MM-ddThh:mm
df.index = df.index.strftime('%Y-%m-%dT%H:%M')
df.to_csv('./data/demand_new.csv')
```
